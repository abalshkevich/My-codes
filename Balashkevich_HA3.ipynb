```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, precision_recall_curve
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_gr3.csv')
x_known  = data["Prediction"]
y_known = data["Correct"]
```

```python
class predictions:
    def __init__(self):
        pass
    def generator(self, p_0):
        y_hat = []
        for p in x_known:
            if p >= p_0:
                y_hat.append(1)
            else:
                y_hat.append(0)
        return y_hat        

    def precision(self, y_hat):
        return precision_score(y_known, y_hat)

    def recall(self, y_hat):
        return recall_score(y_known, y_hat)
        
```

```python
p_zero = np.arange(0, 1.01, 0.01)
y_hatt = [predictions().generator(p) for p in p_zero]
x = [predictions().recall(i) for i in y_hatt]
y = [predictions().precision(i) for i in y_hatt]
```

```python
yy, xx, thresholds = precision_recall_curve(y_known, x_known)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(xx, yy, 'r', linewidth=4, label = 'Sci-kit Learn')
plt.plot(x, y, 'b', linewidth=2, label = 'Manually')
plt.title('Precision-recall curve')
plt.legend()
plt.show()
```
